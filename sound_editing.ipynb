```python
import librosa
import matplotlib.pyplot as plt
```

Importing the librosa library

```python
import numpy as np
import librosa.display
```

```python
print(librosa.__version__)
```

the <b><i>librosa.load("filename")</i> </b>returns two values <b>y, sr</b> they are the <b>audio time series </b>(of np array) and <b>sampling rate</b> of y respectively

```python
y, sr = librosa.core.load("all_sound.mp3")
plt.figure(figsize=(12,8))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.subplot(1, 1, 1)
librosa.display.specshow(D, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
```

## The above spectrogram is of linear frequency from specshow function 

<b><p>where each block representing the planets characteristics</p></b>
<b>Example:</b><i>high speed winds of saturn and hot pressure of venus etc..</i>


```python
y, sr = librosa.core.load("mercury.wav")
plt.figure(figsize=(12,8))
F=librosa.amplitude_to_db(np.abs(librosa.stft(y)),ref=np.max)
plt.subplot(1,1,1)
librosa.display.specshow(F,y_axis="linear")
plt.colorbar(format="%+2.0f dB")
```

## The above spectrogram is of linear frequency of Mercury 

```python
y, sr = librosa.core.load("all_sound.mp3")
plt.figure(figsize=(12,8))
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.subplot(1, 1, 1)
librosa.display.specshow(D, y_axis='log')
plt.colorbar(format='%+2.0f dB')
```

## The above spectrogram is of log frequency from specshow function 

```python
y, sr = librosa.load("all_sound.mp3")
y_harm, y_perc = librosa.effects.hpss(y)
plt.subplot(1, 1, 1)
librosa.display.waveplot(y_harm, sr=sr, alpha=0.5)
librosa.display.waveplot(y_perc, sr=sr, color='r', alpha=0.5)
plt.title('Harmonic + Percussive')
plt.tight_layout()
```

## The above WAVE FORM shows the harmonics and percussive of each planet
